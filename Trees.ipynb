## Data Dictionary 
* https://data.cityofnewyork.us/Environment/2015-Street-Tree-Census-Tree-Data/uvpi-gqnh

1. tree_dbh	Diameter of the tree, measured at approximately 54" / 137cm above the ground
2. curb_loc	Location of tree bed in relationship to the curb
3. steward	Indicates the number of unique signs of stewardship observed for this tree
4. guards	Indicates whether a guard is present, and if the user felt it was a helpful or harmful guard
5. sidewalk	Indicates whether one of the sidewalk flags immediately adjacent to the tree was damaged, cracked, or lifted
6. root_stone	Indicates the presence of a root problem caused by paving stones in tree bed
7. root_grate	Indicates the presence of a root problem caused by metal grates in tree bed
8. root_other	Indicates the presence of other root problems
9. trunk_wire	Indicates the presence of a trunk problem caused by wires or 10. rope wrapped around the trunk
11. trnk_light	Indicates the presence of a trunk problem caused by lighting installed on the tree
12. trnk_other	Indicates the presence of other trunk problems
13. brch_light	Indicates the presence of a branch problem caused by lights (usually string lights) or wires in the branches
14. brch_shoe	Indicates the presence of a branch problem caused by sneakers in the branches
15. brch_other	Indicates the presence of other branch problems
16. postcode	Five-digit zipcode in which tree is located
17. community board	Community board in which tree point is located
borocode	
18. Code for borough in which tree point is located: 1 (Manhattan), 2 (Bronx), 3 (Brooklyn), 4 (Queens), 5 (Staten Island)
19. cncldist	Council district in which tree point is located
20. st_assem	State Assembly District in which tree point is located
21. st_senate	State Senate District in which tree point is located
22. boro_ct	This is the boro_ct identifyer for the census tract that the tree point falls into
23. latitude	Latitude of point, in decimal degrees
24. longitude	Longitude of point, in decimal degrees
25. x_sp	X coordinate, in state plane, in feet
26. y_sp	Y coordinate, in state plane, in feet

```python
# distribution by diameter size of entire dataset

# change type from object to integer for diam measure field 
# create subfram storing only diameters 
diameters = orig['tree_dbh'].astype(int)

# Create axis
fig, ax = plt.subplots(figsize=(40, 4))

# Create histogram of the diameters on axis
ax.hist(diameters, bins=100, color = 'green')
ax.set_xlabel('Diameter (?)')
ax.set_ylabel('Count')
ax.set_title('Histogram of Tree Diameter Distribution')

# Display the plot
plt.show()
```

```python
# box plot to view outliers
# Create axis
fig, ax = plt.subplots(figsize=(2,5))

# Create boxplot of the diameters on axis
ax.boxplot(diameters, labels=['Tree Diameter'])
ax.set_title('Boxplot of Tree Diameter Distribution')

# Display the plot
plt.show()
```

```python
# get dist of diameters by filtersing for diameter values less than 25 - to remove outliers

diameters2 = orig['tree_dbh'].astype(int)
diameters2 = diameters2[diameters2 < 25]

# Create axis
fig, ax = plt.subplots(figsize=(10, 4))

# Create histogram of the diameters on axis
ax.hist(diameters2, bins=50, color='green')

# labels and title 
ax.set_xlabel('Diameter (metric not provided)')
ax.set_ylabel('Count')
ax.set_title('Histogram of Tree Diameter Distribution < 25')

tick_spacing = 5
ax.xaxis.set_major_locator(plt.MultipleLocator(tick_spacing))

# Display the plot
plt.show()
```

```python
# get percentage of diameters of entire dataset 

import numpy as np

# Define the bins with specified range (not equal dist bins)
bins = [0, 2, 4, 6, 8, 10, 15, 20, 25, 30, 35, 40, 45, 50, 100, 200, 300, 400, 500]

# Create histogram using numpy.histogram
counts, _ = np.histogram(diameters, bins=bins)

# Calculate total count
total_count = counts.sum()
#print(total_count)

# Calculate the percentage of count for each bin
percentages = (counts / total_count) * 100

# Print the percentages using enumaerate
for i, percentage in enumerate(percentages):
    if i == 0:
        print("Diameter: {}-{}, Count: {}, Percentage: {:.2f}%".format(bins[i], bins[i+1], counts[i], percentage))
    else:
        print("Diameter: {}-{}, Count: {}, Percentage: {:.2f}%".format(bins[i]+1, bins[i+1], counts[i], percentage))
```

```python
import seaborn as sns

# Create a pivot table with status counts
status_table = orig.pivot_table(index='status', columns='boroname', values='tree_id', aggfunc='count', fill_value=0)

# Create heatmap
sns.heatmap(status_table, cmap='Greens', annot=True, fmt='d')

# Set title and labels
plt.title('Tree Status by Borough')
plt.xlabel('Borough')
plt.ylabel('Status')

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Get top 5 zipcodes by count for each boroname
top_zipcodes = {}
for boro in df["boroname"].unique():
    top_zipcodes[boro] = df[df["boroname"] == boro]["zipcode"].value_counts().head(5).index.tolist()

# Create dictionary of dataframes for each boro and corresponding top zipcodes
df_dict = {}
for boro in df["boroname"].unique():
    df_dict[boro] = df[(df["boroname"] == boro) & (df["zipcode"].isin(top_zipcodes[boro]))]

# Create stacked barplot
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25,6), sharey=True)
for i, boro in enumerate(df["boroname"].unique()):
    sns.countplot(x="zipcode", data=df_dict[boro], hue="boroname", ax=axes[i], palette="Greens")
    axes[i].set_title(f"Count of trees by zipcode (top 5) for {boro}")
    axes[i].legend(loc="upper right")
plt.tight_layout()
plt.show()

```

```python
boro_dfs = {}
for boro in df['boroname'].unique().tolist():
    boro_subframe = df.loc[df['boroname'] == boro].copy()
    boro_dfs[boro] = boro_subframe.groupby('zipcode').size().nlargest(5).reset_index(name='count')
```

```python
boro_dfs
```

```python

# Loop through each borough and create a bar chart
for boro_name, boro_df in boro_dfs.items():
    plt.figure()
    plt.barh(boro_df['zipcode'], boro_df['count'], color='green')
    plt.gca().invert_yaxis() # invert y-axis to show zip codes from top to bottom
    plt.title(f'Top 20 Zip Codes with Most Trees in {boro_name}')
    plt.xlabel('Number of Trees')
    plt.ylabel('Zip Code')
    plt.tight_layout()
    plt.show()
```

```python
# to import data alternate way - still throttles t0 1000

import requests
import pandas as pd

url = 'https://data.cityofnewyork.us/resource/uvpi-gqnh.json'
#https://data.cityofnewyork.us/resource/mdmw-kuen.json
params = {'$limit': 10000}

data = []

# retrieve data in batches using pagination
while True:
    response = requests.get(url, params=params)
    if response.status_code != 200:
        print('Error:', response.status_code)
        break
    batch = response.json()
    data.extend(batch)
    if len(batch) < 10000:
        break
    offset = int(params.get('$offset', 0)) + 10000
    params['$offset'] = str(offset)

# flatten the nested JSON data and create a Pandas DataFrame
getdf = pd.json_normalize(data)

# display the first few rows of the DataFrame
getdf.head()
```
